```python
import pandas as pd
import numpy as np
```

```python
# Load sample data
data = pd.DataFrame({
    'VehicleID': [1,2,3,4,5],
    'Make': ['Toyota','Honda','Ford','Toyota','Honda'],
    'Model': ['Corolla','Civic','Focus','Camry','Accord'],
    'Year': [2018,2020,2015,2019,2016],
    'Mileage': [25000,10000,90000,70000,120000]
})
```

```python
# Calculate vehicle age
current_year = 2025
data['Age'] = current_year - data['Year']

```

```python
# Calculate expected average mileage per year (assuming 12,000 miles/year)
data['ExpectedMileage'] = data['Age'] * 12000

```

```python
# Detect anomaly: mileage > 1.5x expected or < 0.5x expected
data['Anomaly'] = np.where(
    (data['Mileage'] > 1.5 * data['ExpectedMileage']) | 
    (data['Mileage'] < 0.5 * data['ExpectedMileage']),
    True,
    False
)

print(data[['VehicleID','Make','Model','Year','Mileage','ExpectedMileage','Anomaly']])
```
